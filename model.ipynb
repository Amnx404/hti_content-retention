```python
import pandas as pd
from helper import feature_extractor
from datetime import datetime, timedelta
from tqdm import tqdm

def convert_time_format(iso_time):
    try:
        # Parse ISO 8601 time and add 5:30 offset
        time = datetime.fromisoformat(iso_time.replace("Z", "+00:00")) + timedelta(hours=5, minutes=30)
        return time.strftime('%Y-%m-%d %H-%M-%S')
    except Exception as e:
        print(f"Error converting time: {iso_time}, {e}")
        return None

# Define a function to extract and store raw features for each activity segment
def extract_and_store_features(df):
    # List of activity columns with start and end times
    activity_columns = [
        ('word_recall_start_1', 'digit_complete_1'),
        ('word_recall_start_2', 'digit_complete_2'),
        ('youtube_start', 'youtube_end')
        
    ]
    
    # Iterate through each activity type
    for start_col, end_col in tqdm(activity_columns):
        activity_name = start_col  # Use start column name as activity identifier
        feature_col_name = f"{activity_name}_features"  # Name of the new column to store features
        
        # Extract features for each row in the dataframe
        features_list = []
        for _, row in tqdm(df.iterrows()):
            try:
                start_time = convert_time_format(row[start_col])
                end_time = convert_time_format(row[end_col])
                # print(f"Processing {activity_name}: {start_time} to {end_time}")
                
                if start_time and end_time:
                    # Call the feature extractor for the given time range
                    extracted_features = feature_extractor.extract_features(start_time, end_time)
                    features_list.append(extracted_features)  # Store raw extracted features
                else:
                    features_list.append(None)  # Append None if conversion or extraction fails
            except Exception as e:
                print(f"Error processing {activity_name}: {e}")
                features_list.append(None)
                print("processing " + activity_name, start_time, end_time)
        
        # Add the list of features as a new column in the dataframe
        df[feature_col_name] = features_list
    
    return df

# Load your input dataframe (replace 'part' with your dataframe variable or data-loading logic)
# Example: part = pd.read_csv("your_dataset.csv")

# Process the dataframe and store raw features
entire_df = extract_and_store_features(part)
```

```python
final_features = [
        'Positive_Engagement', 'Stress_Score', 'Happiness_Stress_Ratio',
        'Cognitive_Load', 'Engagement_Score', 'Head_Movement_Intensity',
        'Gaze_Consistency', 'Valence', 'Arousal', 'AU_Peak_Ratio'
    ]

for i in range(len(final_features)):
    final_features[i] += '_norm'

final_features.extend(['digit_sc_change', 'word_sc_change','format'])
```

```python
import pandas as pd

# Initialize lists to store processed data
processed_start_1 = []
processed_start_2_diff = []

# Process rows in the specified columns
if 'word_recall_start_1_features' in entire_df.columns and 'word_recall_start_2_features' in entire_df.columns:
    for index, row in entire_df.iterrows():
        # Process `word_recall_start_1_features`
        features_start_1 = row['word_recall_start_1_features']
        if features_start_1 is not None and not features_start_1.empty:
            # Drop unnecessary columns
            features_start_1 = features_start_1.drop(
                columns=['absolute_timestamp', 'Segment Start', 'timestamp'], errors='ignore'
            )
            # Calculate mean for each row
            mean_start_1 = features_start_1.mean()
            renamed_start_1 = {f"word_recall_start_1_features_{key}": value for key, value in mean_start_1.items()}
            processed_start_1.append(renamed_start_1)

            # Process `word_recall_start_2_features` and calculate differences
            features_start_2 = row['word_recall_start_2_features']
            if features_start_2 is not None and not features_start_2.empty:
                # Drop unnecessary columns
                features_start_2 = features_start_2.drop(
                    columns=['absolute_timestamp', 'Segment Start', 'timestamp'], errors='ignore'
                )
                # Calculate the difference between each subrow in column 2 and the mean from column 1
                diff = features_start_2 - mean_start_1
                renamed_diff = {f"{key}_norm": value for key, value in diff.mean().items()}
                processed_start_2_diff.append(renamed_diff)

# Convert processed lists into DataFrames
processed_start_1_df = pd.DataFrame(processed_start_1)
processed_start_2_diff_df = pd.DataFrame(processed_start_2_diff)

# Combine all processed data with the original DataFrame
score_diff = pd.concat(
    [
        entire_df.reset_index(drop=True),
        processed_start_2_diff_df.reset_index(drop=True),
    ],
    axis=1,
)

```
