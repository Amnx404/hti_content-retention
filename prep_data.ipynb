```python
import json
from datetime import datetime
import os
import pandas as pd
import numpy as np
from scipy.stats import linregress

def extract_data_between_timestamps(data, start_timestamp, end_timestamp):
    """
    Extract heart rate data entries from nested JSON data between two timestamps.

    :param data: Dictionary containing nested HR data entries under 'metrics'.
    :param start_timestamp: Start timestamp as a string in the format '%Y-%m-%d %I:%M:%S %p %z'
    :param end_timestamp: End timestamp as a string in the format '%Y-%m-%d %I:%M:%S %p %z'
    :return: Filtered list of HR data entries between the given timestamps.
    """
    # Convert start and end timestamps to datetime objects
    try:
        start_dt = datetime.strptime(start_timestamp, '%Y-%m-%d %I:%M:%S %p %z')
        end_dt = datetime.strptime(end_timestamp, '%Y-%m-%d %I:%M:%S %p %z')
    except ValueError as ve:
        raise ValueError(f"Incorrect timestamp format: {ve}")

    filtered_data = []

    # Access 'metrics' within 'data'
    metrics = data.get('metrics', [])
    if not metrics:
        print("Warning: No 'metrics' key found in the data.")
        return filtered_data

    # Iterate over each metric
    for metric_index, metric in enumerate(metrics):
        metric_data = metric.get('data', [])
        if not metric_data:
            print(f"Warning: No 'data' found for metric at index {metric_index}.")
            continue

        # Iterate over each entry in the metric's data
        for entry_index, entry in enumerate(metric_data):
            # Ensure the entry is a dictionary
            if not isinstance(entry, dict):
                print(f"Warning: Entry at metrics[{metric_index}]['data'][{entry_index}] is not a dictionary. Skipping.")
                continue

            # Extract and parse the 'date' field
            date_str = entry.get("date")
            if not date_str:
                print(f"Warning: Entry at metrics[{metric_index}]['data'][{entry_index}] missing 'date'. Skipping.")
                continue

            try:
                # Normalize the date string by replacing non-breaking spaces and other Unicode spaces
                date_str_normalized = ' '.join(date_str.split())
                entry_dt = datetime.strptime(date_str_normalized, '%Y-%m-%d %I:%M:%S %p %z')
            except ValueError as ve:
                print(f"Warning: Invalid date format in entry at metrics[{metric_index}]['data'][{entry_index}]: {ve}. Skipping.")
                continue

            # Check if the entry's date is within the specified range
            if start_dt <= entry_dt <= end_dt:
                filtered_data.append(entry)

    return filtered_data

def extract_hr_features(hr_data, granularity):
    """
    Extracts relevant heart rate features from HR data based on specified granularity.

    :param hr_data: List of dictionaries containing HR data entries.
    :param granularity: Segment duration in seconds (integer).
    :return: pandas.DataFrame containing extracted features for each segment.
    """
    
    if not hr_data:
        raise ValueError("No HR data provided for feature extraction.")

    # Convert hr_data to DataFrame
    df = pd.DataFrame(hr_data)
    
    # Parse 'date' to datetime using pandas for better handling of various formats
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    
    # Drop rows with invalid dates
    df = df.dropna(subset=['date'])
    
    # Sort by date
    df = df.sort_values('date').reset_index(drop=True)
    
    if df.empty:
        raise ValueError("No valid HR data available after parsing dates.")
    
    # Set 'date' as index
    df.set_index('date', inplace=True)
    
    # Resample data based on granularity
    resample_rule = f'{granularity}S'
    grouped = df.resample(resample_rule)
    
    # Initialize lists to store features
    feature_list = []
    
    # Iterate over each segment
    for segment_time, group in grouped:
        if group.empty:
            # Handle empty segments by assigning NaN
            feature = {
                'Segment Start': segment_time,
                'Mean HR': np.nan,
                'HR Slope': np.nan,
                'HR Variability': np.nan,
                'Peak HR': np.nan,
                'HR Change': np.nan,
                'Moving Average HR': np.nan,
                'HR Transitions': np.nan
            }
        else:
            # Extract HR values
            hr_values = group['Avg'].astype(float).values
            hr_max = group['Max'].astype(float).values
            
            # Mean HR
            mean_hr = np.mean(hr_values)
            
            # HR Slope using linear regression (time in seconds since segment start)
            times = (group.index - segment_time).total_seconds()
            if len(times) > 1:
                slope, intercept, r_value, p_value, std_err = linregress(times, hr_values)
            else:
                slope = 0  # Not enough data points to determine slope
            
            # HR Variability (Standard Deviation)
            hr_variability = np.std(hr_values)
            
            # Peak HR
            peak_hr = np.max(hr_max)
            
            # Moving Average HR (over the segment)
            moving_avg = mean_hr  # Equivalent to mean_hr in this context
            
            # HR Transitions: Number of times HR changes by more than 10 bpm between consecutive entries
            if len(hr_values) > 1:
                transitions = np.sum(np.abs(np.diff(hr_values)) > 10)
            else:
                transitions = 0  # Not enough data points to determine transitions
            
            feature = {
                'Segment Start': segment_time,
                'Mean HR': mean_hr,
                'HR Slope': slope,
                'HR Variability': hr_variability,
                'Peak HR': peak_hr,
                'HR Change': np.nan,  # To be filled later
                'Moving Average HR': moving_avg,
                'HR Transitions': transitions
            }
        
        feature_list.append(feature)
    
    # Create DataFrame from features
    features_df = pd.DataFrame(feature_list)
    
    # Calculate HR Change (difference in Mean HR from previous segment)
    features_df['HR Change'] = features_df['Mean HR'].diff()
    
    # Fill NaN for the first segment's HR Change with 0
    features_df['HR Change'].fillna(0, inplace=True)
    
    # Reset index to have 'Segment Start' as a column
    features_df.reset_index(drop=True, inplace=True)
    
    return features_df

# Sample Usage

if __name__ == "__main__":
    # Define the correct file path
    file_path = './Heart Rate.json'
    
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file '{file_path}' was not found in the environment.")
    
    # Load JSON data from the file
    with open(file_path, 'r', encoding='utf-8') as file:
        try:
            json_data = json.load(file)
        except json.JSONDecodeError as jde:
            raise ValueError(f"Error decoding JSON: {jde}")
    
    # Inspect the structure of json_data['data']
    print(f"Type of json_data['data']: {type(json_data['data'])}")

    if isinstance(json_data['data'], dict):
        print(f"Keys in json_data['data']: {json_data['data'].keys()}")
        # Assuming 'metrics' is the key containing the list of metrics
        metrics = json_data['data'].get('metrics', [])
        if metrics:
            first_metric = metrics[0]
            print(f"Sample metric (metrics[0]): {first_metric}")
            sample_entry = first_metric.get('data', [])[0] if first_metric.get('data') else None
            print(f"Sample entry in first metric (metrics[0]['data'][0]): {sample_entry}")
        else:
            print("No metrics found in 'data'.")
    elif isinstance(json_data['data'], list):
        print(f"Number of entries in json_data['data']: {len(json_data['data'])}")
        if json_data['data']:
            print(f"Sample entry: {json_data['data'][0]}")
    else:
        print("json_data['data'] is neither a list nor a dict.")
    
    # Define start and end times
    start_time = '2024-11-10 12:00:00 AM +0530'
    end_time = '2024-12-14 12:30:00 AM +0530'
    
    # Define granularity in seconds
    granularity = 60  # 1-minute segments
    
    # Extract filtered data using the updated extract_data_between_timestamps function
    if isinstance(json_data['data'], dict):
        # Access 'metrics' list
        metrics = json_data['data'].get('metrics', [])
        if not metrics:
            raise ValueError("No 'metrics' found in 'data'.")
        
        # Extract data from all metrics
        filtered_entries = extract_data_between_timestamps(json_data['data'], start_time, end_time)
    elif isinstance(json_data['data'], list):
        # If 'data' is a list, process it directly
        filtered_entries = extract_data_between_timestamps(json_data['data'], start_time, end_time)
    else:
        raise ValueError("Unsupported 'data' structure in JSON.")
    
    # Check if any data was filtered
    if not filtered_entries:
        print("No HR data found within the specified time range.")
    else:
        # Extract features using the extract_hr_features function
        try:
            features = extract_hr_features(filtered_entries, granularity)
            # Display the extracted features
            print(features)
        except ValueError as ve:
            print(f"Error during feature extraction: {ve}")
```

```python
# Feature: Segment Start
# Description: The starting timestamp of each defined time segment.
# Implication: Marks the beginning of a period for which heart rate features are calculated, allowing temporal analysis.

# Feature: Mean HR
# Description: The average heart rate within the segment.
# Implication: Provides a general overview of heart rate levels during the segment; higher values may indicate increased activity or stress.

# Feature: Median HR
# Description: The median heart rate value within the segment.
# Implication: Offers a central tendency measure less affected by outliers, ensuring a robust representation of typical heart rate.

# Feature: HR Slope
# Description: The trend of heart rate over time within the segment, calculated using linear regression.
# Implication: Indicates whether heart rate is increasing, decreasing, or stable; useful for detecting rising stress or calming trends.

# Feature: HR Variability
# Description: The standard deviation of heart rate values within the segment.
# Implication: Reflects the consistency of heart rate; higher variability may suggest fluctuating activity levels or stress.

# Feature: Coefficient of Variation
# Description: The ratio of HR Variability to Mean HR within the segment.
# Implication: Normalizes variability relative to the mean, allowing comparison across segments with different average heart rates.

# Feature: Peak HR
# Description: The maximum heart rate observed within the segment.
# Implication: Identifies the highest intensity of heart rate during the segment, potentially indicating moments of exertion or stress.

# Feature: HR Change
# Description: The difference in Mean HR compared to the previous segment.
# Implication: Highlights shifts in heart rate trends, such as sudden increases or decreases, which may correspond to changes in activity or emotional state.

# Feature: Moving Average HR
# Description: The rolling average of Mean HR over a specified number of consecutive segments.
# Implication: Smooths out short-term fluctuations to reveal underlying trends in heart rate over time.

# Feature: HR Transitions
# Description: The number of significant heart rate changes (e.g., >10 bpm) between consecutive data points within the segment.
# Implication: Indicates the frequency of abrupt heart rate shifts, which can signal stress, intense activity, or physiological irregularities.
```

```python
import json
from datetime import datetime, timedelta
import os
import pandas as pd

def extract_video_data_between_timestamps(video_files, start_timestamp, end_timestamp):
    """
    Extract data entries from a series of video files between two timestamps.

    :param video_files: Dictionary of video file names and their durations in seconds
                       {"video_name.csv": duration_in_seconds}
    :param start_timestamp: Start timestamp as a string in the format '%Y-%m-%d %H-%M-%S'
    :param end_timestamp: End timestamp as a string in the format '%Y-%m-%d %H-%M-%S'
    :return: Dictionary with video file names and corresponding filtered DataFrames
    """
    start_dt = datetime.strptime(start_timestamp, '%Y-%m-%d %H-%M-%S')
    end_dt = datetime.strptime(end_timestamp, '%Y-%m-%d %H-%M-%S')

    filtered_data = {}

    for video_name, duration in video_files.items():
        # Extract the start time from the video name
        video_start_str = video_name.split('.')[0]  # Remove file extension
        video_start_dt = datetime.strptime(video_start_str, '%Y-%m-%d%H-%M-%S')
        video_end_dt = video_start_dt + timedelta(seconds=duration)

        # Check if the video overlaps with the desired time range
        if video_end_dt < start_dt or video_start_dt > end_dt:
            continue  # Skip videos outside the range

        # Load the video data
        video_path = f'./Out/{video_name}'
        if not os.path.exists(video_path):
            raise FileNotFoundError(f"The video file '{video_name}' was not found.")

        video_data = pd.read_csv(video_path)

        # If timestamp is in seconds relative to video start, use it directly
        video_data['absolute_timestamp'] = video_start_dt + pd.to_timedelta(video_data[' timestamp'], unit='s')

        # Filter rows within the desired time range
        filtered_video_data = video_data[(video_data['absolute_timestamp'] >= start_dt) &
                                         (video_data['absolute_timestamp'] <= end_dt)]

        filtered_data[video_name] = filtered_video_data

    return filtered_data

# Example usage:
video_files = video_duration_dict

start_time = '2024-12-01 22-55-00'
end_time = '2024-12-01 23-02-00'

filtered_videos = extract_video_data_between_timestamps(video_files, start_time, end_time)

for video_name, data in filtered_videos.items():
    print(f"Filtered data for {video_name}:")
    print(data.head())
    print(data.tail())
```

```python
import pandas as pd
import json

def process_session_durations_to_json(csv_file_path, output_file_path):
    # Load the CSV file
    data = pd.read_csv(csv_file_path)
    
    # Ensure the relevant columns are present
    required_columns = ['word_recall_start_1', 'digit_complete_2']
    for col in required_columns:
        if col not in data.columns:
            raise KeyError(f"Column '{col}' is missing from the CSV file.")
    
    # Convert timestamps to datetime format
    data['word_recall_start_1'] = pd.to_datetime(data['word_recall_start_1'], errors='coerce')
    data['digit_complete_2'] = pd.to_datetime(data['digit_complete_2'], errors='coerce')
    
    # Assign start and end times
    data['start_time'] = data['word_recall_start_1']
    data['end_time'] = data['digit_complete_2']
    
    # Format timestamps for the desired output
    data['formatted_start_time'] = data['start_time'].dt.strftime('%Y-%m-%d %H-%M-%S')
    data['formatted_end_time'] = data['end_time'].dt.strftime('%Y-%m-%d %H-%M-%S')
    
    # Create the output list
    formatted_durations = [
        {
            'start_time': row['formatted_start_time'],
            'end_time': row['formatted_end_time']
        }
        for _, row in data.iterrows()
    ]
    
    # Save as JSON
    with open(output_file_path, 'w') as output_file:
        json.dump(formatted_durations, output_file, indent=4)
    
    print(f"Data saved to {output_file_path}")

# Example usage
csv_file_path = 'memory_tests_rows.csv'  # Replace with your actual file path
output_file_path = 'testduration_data.json'  # Replace with your desired output file path
process_session_durations_to_json(csv_file_path, output_file_path)
```
